```python
def format_string(input_string):
    # Add a new line before every "Question:"
    formatted_string = re.sub(r'(Question:|Answer:)', r'\n\1', input_string)
    # Add a new line after every "Question:" or "Answer:"
    formatted_string = re.sub(r'(Question:|Answer:)', r'\1\n', formatted_string)
    return formatted_string

print(format_string(df['Match Support Contact Notes'][7])) #  'going really well', 'a great kid'
```
